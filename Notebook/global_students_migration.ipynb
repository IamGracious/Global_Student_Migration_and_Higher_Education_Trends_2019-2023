```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Loading the dataset
df = pd.read_csv(r"C:\Users\User\Downloads\global_student_migration.csv")
p
# Basic dataset inspection
print(df.head())
print(df.info())
print(df.describe())
print(df.columns)
df.columns = df.columns.str.capitalize()
```

```python
# Check for missing values
print(df.isnull().sum())

# Example: Fill or drop missing values
df = df.dropna()  # Or use df.fillna(value)
```

```python
plt.figure(figsize=(10,6))
sns.countplot(y='destination_country', 
              data=df, 
              order=df['destination_country'].value_counts().index)
plt.title("Number of Students per Destination Country (2019-2023)")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(
    y='Course_name', 
    data=df, 
    order=df['Course_name'].value_counts().index[:10]
)
plt.title("Top 10 Courses Studied by International Students")
plt.show()
```

```python
sns.countplot(x='Scholarship_received', data=df)
plt.title("Scholarship Distribution Among Students")
plt.show()
```

```python
print(df.columns)
```

```python
# Average salary by destination country
avg_salary = df.groupby('Destination_country')['Starting_salary_usd'].mean().sort_values(ascending=False)
print(avg_salary)

# Placement rate per country
placement_rate = df.groupby('Destination_country')['Placement_status'].value_counts(normalize=True).unstack()
print(placement_rate)
```

```python
migration_matrix = df.groupby(['Origin_country', 'Destination_country']).size().unstack(fill_value=0)
print(migration_matrix)
```

```python
time_trend = df.groupby(['Year_of_enrollment', 'Destination_country']).size().unstack(fill_value=0)
time_trend.plot(kind='line', figsize=(12,6))
plt.title("Student Migration Trends by Year")
plt.ylabel("Number of Students")
plt.show()
```
